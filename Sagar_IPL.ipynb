```python
#Extracting year value from date colum

match_data['season']= pd.DatetimeIndex(match_data['date']).year

match_data.head()
```

```python
#Total number of matches per season
match_per_season= match_data.groupby(['season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season.head(15)
```

```python
match_per_season= match_data.groupby(['season'])['id'].count().reset_index()
```

```python
match_per_season.head(200)
```

```python
#plotting bar graph using seasoborn library

sns.countplot(match_data['season'])
#using matplot library

plt.xticks(rotation= 40, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Season',fontsize=10)
plt.ylabel('Matches',fontsize=10)
plt.title('Total Matches played in each Season',fontsize=15,fontweight="bold")


```

```python
sns.lineplot(data=match_per_season,x='season',y='matches')
plt.title('Total Matches played in each Season',fontsize=15,fontweight="bold")

plt.show()
```

```python
season_data= match_data[['id','season']].merge(ball_data, left_on='id', right_on='id', how ='left').drop('id',axis=1)
season_data.head()
```

```python
#plotting line graph

season=season_data.groupby(['season'])['total_runs'].sum().reset_index()
p=season.set_index('season')
#ax.set(facecolor="grey")
#ax=plt.axes()
sns.lineplot(data=p, palette='magma')
plt.title('Total number of run scored each season', fontsize=15, fontweight='bold')
plt.show()

season.head(15)
```

```python
season.head()
```

```python
season.head()
```

```python
match_per_season.head()
```

```python
runs_per_season= pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs_per_season['Average runs per Match']=runs_per_season['total_runs']/runs_per_season['matches']
#runs_per_season.set_index('season',inplace=True)
#runs_per_season.head(100)

sns.lineplot(data=runs_per_season,x='season',y='Average runs per Match')
plt.title('\n Average runs scored per season',fontweight='bold',fontsize=15)
plt.show()

```
